```python
import os
# S3  
s3_client = boto3.client('s3')

def download_files_from_s3(bucket_name, prefix, local_dir):
    #      
    response = s3_client.list_objects_v2(Bucket=bucket_name, Prefix=prefix)

    #   
    if 'Contents' in response:
        for obj in response['Contents']:
            file_key = obj['Key']
            #   
            file_name = os.path.basename(file_key)
            
            #     
            local_file_path = os.path.join(local_dir, file_name)

            print(f"Downloading {file_key} to {local_file_path}...")
            
            # S3   
            s3_client.download_file(bucket_name, file_key, local_file_path)
        print("Download complete.")
    else:
        print("No files found in the specified S3 folder.")


controller_name = "V2LC"
main_test_name = "HEV_P2_ACOverLoad_IG1_1"
sub_test_name = "Severe1_Above2.83s_1"


#  
bucket_name = 'kefico-source'
prefix = f'main_parquet_dir/{controller_name}/{main_test_name}/{sub_test_name}/'  #   (: 'folder/subfolder/')
local_dir = 'opt/ml/processing/input/raw'  #   

download_files_from_s3(bucket_name, prefix, local_dir)
```
